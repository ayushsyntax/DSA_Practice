```python
#Check if a Number is Palindrome or Not
n = 1234
nums = n
result = 0
while nums > 0:
  last = nums % 10
  result = result * 10 + last
  nums = nums // 10
print(result == n)
```

```python
#Armstrong Number
n = 153
nums = n
result = 0
count = 0
while nums > 0:
  count+=1
  nums = nums//10
nums = n
while nums > 0:
  last = nums % 10
  result = result + last ** count
  nums = nums // 10
print(result == n)
```

```python
# Armstrong Number
n = 153
nums = n
number_of_digits = len(str(nums))
result = 0

while nums > 0:
    last = nums % 10
    result += last ** number_of_digits
    nums = nums // 10

print(result == n)
```

```python
#Print All Factors of a Given Number
from math import sqrt
#optimal solution t.c o(sqrt(n)) s.c = o(k)
def print_factors(n):
    result = []
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:
            result.append(i)
            if i != n // i:
                result.append(n // i)
    result.sort()
    return result

n = 18
print(print_factors(n))
```

```python
#brute force
#t.c = o(n) s.c = o(k)
def print_factors(n):
    result = []
    for i in range(1, n + 1):
        if n % i == 0:
            result.append(i)
    return result
print(print_factors(n))
```

```python
#better approch
#t.c nearly o(n) or o(n/2) s.c = o(k)
def print_factors(n):
    result = []
    for i in range(1, (n//2)+1):
        if n % i == 0:
            result.append(i)
    result.append(n)
    return result
print(print_factors(n))
```

```python
#Store Frequency in Dictionary
#Method 1 T.C = o(n) s.c = o(n)
def store_frequency(nums):
    hash_map = dict()
    for i in range(len(nums)):
        if nums[i] in hash_map:
            hash_map[nums[i]] += 1
        else:
            hash_map[nums[i]] = 1
    return hash_map

nums = [1,2,3,4,5,6,33,44,33,44,2,1,2,1]
x = store_frequency(nums)
x[1]
```

```python
#Method 2 T.C= o(n) S.C = o(n)
def store_frequency(nums):
    hash_map = {}
    n = len(nums)
    for i in range(n):
        hash_map[nums[i]] = hash_map.get(nums[i], 0) + 1
    return hash_map
nums = [3,2,3,4,5,6,33,44,33,44,2,1,2,1,4]
x = store_frequency(nums)
x[4]
```

```python
#functional recurison
def recursion(N):
    if N == 1:
        return 1
    return N + recursion(N-1)
print(recursion(10))
```

```python
def recursion(N):
    if N==0 or N==1:
        return 1
    return N * recursion(N - 1)

print(recursion(5))
```

```python
#reverse an array
nums = [2, 4, 5, 21, 3, 54]

def recursion(nums, L, R):
    if L >= R:
        return nums
    nums[L], nums[R] = nums[R], nums[L]
    return recursion(nums, L + 1, R - 1)

reversed_array = recursion(nums, 3, len(nums) - 1)
print(reversed_array)

#Time Complexity: O(n)
#Space Complexity: O(n) (due to the recursion stack)
```

Since an empty string reads the same forward and backward, it is a palindrome.


```python
class Solution(object):
    def isPalindrome(self, s):
        """
        :type s: str
        :rtype: bool
        """

        filtered = []
        for char in s:
            if char.isalnum():
                filtered.append(char.lower())
        l, r = 0, len(filtered) - 1
        while l<r:
            if filtered[l]!= filtered[r]:
                return False
            l +=1
            r -= 1
        return True

        #t.c = o(n)
        #s.c = o(n)
```

```python
class Solution(object):
    def isPalindrome(self, s):
        l, r = 0, len(s) - 1

        while l < r:
            while l < r and not s[l].isalnum():
                l += 1
            while l < r and not s[r].isalnum():
                r -= 1

            if s[l].lower() != s[r].lower():
                return False

            l += 1
            r -= 1

        return True
#t.c = o(n)
#s.c = o(1)
```

```python
#selection sort
#ascending
n  = [1,23,42,2,44,2,1,15]
def selectionsort(n):
  l = len(n)
  for i in range(0,l):
    mini_index = i
    for j in range(i+1, l):
      if n[mini_index] > n[j]:
        mini_index = j
    n[i],n[mini_index] = n[mini_index],n[i]
  return n
print(selectionsort(n))
#t.c = o(n**2)
#s.c = o(1)
```

```python
#descending
def selectionsort(n):
  for i in range(0,len(n)):
    max_index = i
    for j in range(i+1,len(n)):
      if n[max_index] < n[j]:
        max_index = j
      n[i],n[max_index] = n[max_index],n[i]
  return n
print(selectionsort(n))


```
